```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
```

```python
company=pd.read_csv('Company_Data.csv')
```

```python
company
```

```python
company['highsales'] = np.where((company['Sales']<9), 'Low', 'High')

company
```

```python
comp2 = company.iloc[:,1:]

comp2
```

```python
label_encoder=preprocessing.LabelEncoder()
comp2['ShelveLoc']=label_encoder.fit_transform(comp2['ShelveLoc'])
comp2['Urban']=label_encoder.fit_transform(comp2['Urban'])
comp2['US']=label_encoder.fit_transform(comp2['US'])
comp2['highsales']=label_encoder.fit_transform(comp2['highsales'])
```

```python
comp2
```

```python
x = comp2.iloc[:,:-1]
y = comp2.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#TRAINIG AND TESTING
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=90)
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
predict
```

```python
pd.Series(predict).value_counts()
```

```python
pd.crosstab(y_test,predict)
```
